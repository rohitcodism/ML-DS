```python
def priceRating(x):
    if(x<=10):
        return '$'
    elif(x>10 and x<= 20):
        return '$$'
    elif(x>20 and x <= 30):
        return '$$$'
    elif(x>30 and x<=40):
        return '$$$$'
    else:
        return '$$$$$'
```

```python
df['prRating'] = df['total_bill'].apply(priceRating)
```

```python
def Gen(total_bill, tip):
    ratio = tip/total_bill
    if ratio <= 0.25:
        return 'Avarage'
    else:
        return 'Generous'
```

```python
Gen(16.69, 1.66)
```

```python
df['tip_rating'] = df[['total_bill', 'tip']].apply(lambda df: Gen(df['total_bill'], df['tip']),axis=1)
```

```python
df.head(10)
```

```python
df['tip_rating'] = np.vectorize(Gen)(df['total_bill'], df['tip'])
```

```python
import timeit
```

```python
setup = '''
import numpy as np
import pandas as pd
df = pd.read_csv('tips.csv')
def Gen(total_bill, tip):
    ratio = tip/total_bill
    if ratio <= 0.25:
        return 'Avarage'
    else:
        return 'Generous'
'''
```

```python
st_1 = '''
df['tip_rating'] = df[['total_bill', 'tip']].apply(lambda df: Gen(df['total_bill'], df['tip']),axis=1)
'''

st_2 = '''
df['tip_rating'] = np.vectorize(Gen)(df['total_bill'], df['tip'])
'''
```

```python
timeit.timeit(setup=setup, stmt=st_1, number=1000)
```

```python
timeit.timeit(setup=setup, stmt=st_2, number=1000)
```
