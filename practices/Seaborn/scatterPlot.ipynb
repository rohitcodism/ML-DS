```python
df7 = pd.read_csv('../ML_NoteBook/05-Seaborn/StudentsPerformance.csv')
```

```python
df7
```

```python
sns.boxplot(data=df7, y='math score')
```

```python
plt.figure(figsize=(4,4),dpi=100)
sns.boxplot(data=df7, y='math score', x='test preparation course')
plt.show()
```

```python
plt.figure(figsize=(5,5), dpi=100)
sns.boxplot(data=df7, y='parental level of education', x='reading score', hue='test preparation course', palette='Set2')
plt.legend(bbox_to_anchor = (1.1, 1))
plt.show()
```

```python
plt.figure(figsize=(5,5), dpi=100)
sns.violinplot(data=df7, y='parental level of education', x='reading score', hue='test preparation course', palette='Set2', split=True, inner='quartile')
plt.legend(bbox_to_anchor = (1.1, 1))
plt.show()
```

```python
plt.figure(figsize=(8,4), dpi=100)
sns.swarmplot(data=df7, x='math score', y='gender', size=2, hue='test preparation course', dodge=True)
plt.legend(bbox_to_anchor = (1.5, 1))
```

```python
sns.boxenplot(data=df7, x='math score', y='test preparation course', hue='gender')
```

```python
sns.jointplot(data=df7, x='math score', y='reading score', hue='gender')
```

```python
sns.pairplot(data=df7, hue='gender', corner=True, kind='kde')
```

```python
sns.catplot(data=df7, x='math score', y='gender', kind='swarm', col='lunch', row='test preparation course')
```

```python
g = sns.PairGrid(data=df7, hue='gender')
g = g.map_upper(sns.scatterplot)
g = g.map_lower(sns.kdeplot)
g = g.map_diag(sns.kdeplot, shade=True)
g = g.add_legend()
plt.legend(bbox_to_anchor = (1.5, 1))
```

```python
g = sns.FacetGrid(data=df7, row='gender', col='lunch', hue='gender')
g = g.map(sns.histplot, 'math score', 'reading score')
```
