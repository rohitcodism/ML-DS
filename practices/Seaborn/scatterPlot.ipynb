```python
df = pd.read_csv("../ML_NoteBook/05-Seaborn/dm_office_sales.csv")
```

```python
df
```

```python
plt.figure(figsize=(5,8), dpi=100)
sns.rugplot(x="salary", data=df, height=0.4)
plt.show()
```

```python
#distplot is deprecated , dont use it.

sns.set_style(style='ticks')
sns.displot(data=df, x='salary', linewidth=2, kde=True, rug=True)
```

```python
sns.histplot(data=df, x='salary', kde=True)
plt.show()
```

```python
sns.kdeplot(data=df, x='salary')
plt.show()
```
