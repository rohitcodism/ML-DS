```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('../ML_NoteBook/05-Seaborn/dm_office_sales.csv')
```

```python
df
```

```python
plt.figure(figsize=(5,5), dpi=100)
sns.scatterplot(x="salary", y="sales", data=df, hue="level of education", palette="tab20c", legend=False, size="work experience", alpha=1, style="level of education")
plt.show()

plt.savefig('my_plot.png')
```

Distribution Plot

```python
df = pd.read_csv("../ML_NoteBook/05-Seaborn/dm_office_sales.csv")
```

```python
df
```

```python
plt.figure(figsize=(5,8), dpi=100)
sns.rugplot(x="salary", data=df, height=0.4)
plt.show()
```

```python
#distplot is deprecated , dont use it.

sns.set_style(style='ticks')
sns.displot(data=df, x='salary', linewidth=2, kde=True, rug=True)
```

```python
sns.histplot(data=df, x='salary', kde=True)
plt.show()
```

```python
sns.kdeplot(data=df, x='salary')
plt.show()
```

```python
np.random.seed(42)
sample_ages = np.random.randint(0, 100, 200)
```

```python
sample_ages
```

```python
sample_ages = pd.DataFrame(sample_ages, columns=['age'])
```

```python
sample_ages
```

```python
sns.rugplot(data=sample_ages, x='age')
plt.show()
```

```python
sns.displot(data=sample_ages, x='age', rug=True, bins=60, kde=True)
```

```python
sns.kdeplot(data=sample_ages, x='age', clip=[0,100], bw_adjust=0.5, shade=True)
```
