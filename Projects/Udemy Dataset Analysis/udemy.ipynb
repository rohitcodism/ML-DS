# Udemy Course Data Analysis Report

## Introduction

In this report, we present the findings of a comprehensive data analysis conducted on Udemy courses. The dataset encompassed a wide range of courses from various disciplines available on the Udemy platform. The analysis aimed to uncover insights regarding course categories, pricing, enrollment, and ratings.

## Methodology

The analysis was performed on a dataset containing information about Udemy courses. The dataset included attributes such as course title, instructor, subject category, price, number of subscribers, and average rating. The following steps were taken to derive meaningful insights:

1. **Data Cleaning and Preparation:** The initial dataset was reviewed for inconsistencies, missing values, and outliers. Data cleaning procedures were applied to ensure the accuracy and integrity of the analysis.

2. **Descriptive Statistics:** Descriptive statistics, including mean, median, and standard deviation, were computed for relevant variables such as course price, number of subscribers, and average rating. These statistics provided an overview of the dataset's central tendencies and dispersions.

3. **Category Distribution:** The distribution of courses across different subject categories was examined. This analysis aimed to identify which categories had the highest number of courses and which were relatively less represented.

4. **Price Analysis:** The distribution of course prices was explored, including the range, average, and median prices. Additionally, the relationship between course prices and average ratings was investigated.

5. **Enrollment Analysis:** The distribution of the number of subscribers (enrollments) in courses was analyzed. This helped identify the most popular courses in terms of enrollments.

6. **Rating Analysis:** The distribution of average ratings for courses was examined. The correlation between course ratings and enrollment was also investigated to determine whether higher ratings corresponded to more enrollments.

## Key Findings

The analysis yielded several noteworthy findings:

- **Category Distribution:** The dataset covered a diverse range of subject categories, with categories like "Development," "Business," and  having the highest number of courses. Niche categories such as "Musical Instruments" and "Graphic Design" had relatively fewer courses.

- **Price Analysis:** The majority of courses were priced between $20 and $100, with an average price of approximately $50. Interestingly, courses priced at higher levels did not necessarily correlate with higher average ratings.

- **Subscriber Analysis:** Several courses stood out with exceptionally high subscribers, suggesting that course popularity was not solely dependent on subject category. Instructor reputation and course content likely played significant roles.
