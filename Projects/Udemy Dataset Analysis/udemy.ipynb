##### Importing necessary libraries

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings

warnings.filterwarnings('ignore')
```

```python
dataset = pd.read_csv('./udemydataseet/udemy_courses.csv', parse_dates=['published_timestamp'])
```

```python
dataset = dataset.drop_duplicates()
```

```python
dataset['subject'].value_counts()
```

```python
plt.figure(figsize=(14,5), dpi=150)
sns.countplot(data=dataset, x='subject', palette='magma')
plt.xlabel('Subjects')
plt.ylabel('No. of courses per subject')
plt.show()
```

```python
dataset['level'].value_counts()
```

```python
plt.figure(figsize=(14,5), dpi=150)
sns.countplot(dataset['level'])
plt.xlabel('Levels')
plt.ylabel('Courses per level')
plt.show()
```

```python
dataset['is_paid'].value_counts()
```

```python
plt.figure(figsize=(14,5), dpi=150)
sns.countplot(dataset['is_paid'], palette='crest')
plt.xlabel('Paid')
plt.ylabel('Number of courses')
plt.show()
```

```python
no_lec = dataset.groupby(['is_paid']).sum()
```

```python
no_lec
```

```python
no_lec = no_lec.drop('course_id', axis=1)
```

```python
plt.figure(figsize=(14,5), dpi=150)
sns.barplot(x='is_paid', y='num_subscribers', data=dataset)
plt.xlabel('Paid')
plt.ylabel('Number of Subscribers')
plt.show()
```

```python
plt.figure(figsize=(14,5), dpi=150)
sns.barplot(x='level', y='num_subscribers', data=dataset)
plt.xlabel('Levels')
plt.ylabel('Number of Subscribers')
plt.show()
```

```python
dataset[dataset.num_subscribers == dataset.num_subscribers.max()]['course_title']
```

```python
top_10_courses = dataset.sort_values(by='num_subscribers', ascending=False).head(10)
```

```python
plt.figure(figsize=(14,5), dpi=150)
sns.barplot(x='num_subscribers', y='course_title', data=top_10_courses)
plt.xlabel('Name of the Course')
plt.ylabel('Number of subscribers')
plt.show()
```

```python
dataset[dataset.num_reviews == dataset.num_reviews.max()]['course_title']
```

```python
price_effect = dataset.groupby(['price']).mean()
```

```python
price_effect
```

```python
price_effect = price_effect.drop(['course_id', 'content_duration'], axis=1)
```

```python
price_effect
```

```python
plt.figure(figsize=(14,5), dpi=150)
sns.lineplot(data=price_effect, x='price', y='num_reviews')
plt.xlabel('Price Of the course')
plt.ylabel('Number Of reviews')
plt.show()
```
